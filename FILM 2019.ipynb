# Timeline

```python
from datetime import datetime
```

```python
def modify_dataframe(df):
    """ Modify dataframe to include new columns """
    df['Month'] = pd.to_datetime(df['Date'], format='%Y-%m-%d').dt.month
    return df
```

```python
#The pd.to_datetime function converts the series df['Date'] to a datetime series; and I'm creating a new column called Month 
#which equates to the month number.
#From this column, I can generate X and Y coordinates for my plot.
```

```python
def get_x_y(df):
    """ Get X and Y coordinates; return tuple """
    series = df['Month'].value_counts().sort_index()
    new_series = series.reindex(range(1,13)).fillna(0).astype(int)
    return new_series.index, new_series.values
```

```python
#This takes in my modified dataframe and creates a series that counts the number of occurrences of each month. 
#Then if there are any missing months, fillna fills them in with a value of 0.
```

```python
df = modify_dataframe(df)
X, Y = get_x_y(df)
```

```python
fig = plt.figure(figsize=(14,5))
ax = fig.add_subplot(1, 1, 1)
xlabels = [datetime(2019, i, 1).strftime("%B") for i in list(range(1,13))]
ax.set_xticklabels(xlabels, rotation=45, ha='right')
ax.set_xticks(range(1,13))
ax.set_title('Films Watched Per Month', fontsize=20)
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.plot(X, Y, color="blue", linewidth=2, marker="o")
```

For the first 4 months of the year I was consistently watching between 10-12 films a month. 

Naturally there is a drop-off in the Summer and I'm pleased to see there is evidence I spent time outside.

Towards the end of September I started watching **Mad Men** so that took up a lot of my viewing time. 

There is a large spike in December as I made a push to reach the 100 mark, falling just short at 99.
